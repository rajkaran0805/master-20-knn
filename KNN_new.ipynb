```python
import pandas as pd
import numpy as np
import math
```

```python
data=pd.read_csv("C:/Users/rajka/Desktop/20.1/nba_2013.csv")
print(data.columns.values)
```

```python
data.isnull().any()
```

```python
data.fillna(data.mean(),inplace= True)
```

```python
data.isnull().any()
```

```python
distance_columns = ['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.', 'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']
data_numeric = data[distance_columns]
data = (data_numeric - data_numeric.mean()) / data_numeric.std()
```

```python
import random
from numpy.random import permutation
random_indices = permutation(data.index)
test_cutoff = math.floor(len(data)/3)
test = data.loc[random_indices[1:test_cutoff]]
train = data.loc[random_indices[test_cutoff:]]
```

```python
x_columns = ['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.', 'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf']
y_column = ["pts"]
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=21)
knn.fit(train[x_columns], train[y_column])
predictions = knn.predict(test[x_columns])
```

```python
actual = test[y_column]
mse = (((predictions - actual) ** 2).sum()) / len(predictions)
```
